this is new model, designed to account for ada 

```python
ada = create_model("ada", fold = 5, round = 2) 
```

```python
print(ada)
```

```python
rf = create_model("rf",fold = 5, round = 2)
```

```python
qda = create_model("qda",fold = 5, round = 2)
```

before tuning the machine: ada (78%), rf (83%), qda (46%)

```python
tuned_ada = tune_model(ada, optimize='AUC', round = 2)
print(tuned_ada)
```

```python
tuned_rf = tune_model(rf, optimize='AUC', round = 2)
print(tuned_rf)
```

```python
tuned_qda = tune_model(qda, optimize='AUC', round = 2)
print(tuned_qda)
```

After tunning model: ada, rf, qda (83%) 

```python
plot_model(tuned_rf, plot = 'parameter')
```

```python
evaluate_model(tuned_rf)
```

```python
predict_model(tuned_rf).style.background_gradient(cmap='Greys')
```

```python
final_rf= finalize_model(tuned_rf)
```
